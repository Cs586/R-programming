# All functions used are below

```r
# MTM

calc_mtm = function(MTM,shares,universe,cur_day,cur_stock,cash_account){
    for (i in 1:length(cur_stock)) {
        MTM[i] = shares[i] * universe[cur_day,paste(cur_stock[i],'_Close',sep="")]
    }
return(sum(MTM)+cash_account)
}
```

```r
# finding 5 stock which are surged most or droped most


adj.close_change = function(stock,cur_day,days_interval,universe){
    change = 0
    for (i in 1:length(stock)){
        change[i] = universe[cur_day,paste(stock[i],'_Adj.Close',sep="")]/universe[cur_day-days_interval,paste(stock[i],'_Adj.Close',sep="")]
    }
    new_data = cbind(stock = stock, change = change)
    sorted_new_data = new_data[order(new_data[,2], decreasing = FALSE), ]
    high = sorted_new_data[6:10,]
    low = sorted_new_data[1:5,]
#    return(high)
    return(low[,1])
}

# Test
adj.close_change(stock,6,days_interval[5],universe)
```

```r
# update cash_account after buying the stock, only used once before while loop


initial_buy = function(cur_stock,cur_day,shares,universe,cash_account){
    total_invested = 0
    for (i in 1:length(cur_stock)){
        total_invested[i] = shares[i] * universe[cur_day,paste(cur_stock[i],'_Close',sep="")]
    }
    cash_account = cash_account - sum(total_invested)
    return(cash_account)
}
```

```r
# update cash_account after selling the stock

sell = function(cur_stock,cash_account,cur_day,shares,universe){
    total_cashed = 0
    for (i in 1:length(cur_stock)){
        total_cashed[i] = shares[i] * universe[cur_day,paste(cur_stock[i],'_Close',sep="")]
    }
    cash_account = cash_account + sum(total_cashed)
    return(cash_account)
}
```

```r
# update cur_day

day_change = function(days_interval,cur_day){
    cur_day = cur_day + days_interval
    return(cur_day)
}
```

```r
# update cash_account after buying stock

buy_stock = function(cur_stock,universe,cur_day,shares,cash_account){
    total_invested = 0
    for (i in 1:length(cur_stock)){
        total_invested[i] = shares[i] * universe[cur_day,paste(cur_stock[i],'_Close',sep="")]
    }
    cash_account = cash_account - sum(total_invested)
    return(cash_account)
}
```

```r
# update shares when rebalancing

num_shares = function(cur_stock,cash_account,universe,cur_day){
    shares = 0
    for (i in 1:length(cur_stock)){
        shares[i] = floor(cash_account/5/universe[cur_day,paste(cur_stock[i],'_Close',sep="")])
    }
    return(shares)
}
```

```r
# update dividend value 

get_dividend = function(universe,cur_stock,cur_day,shares){
    dividend = 0
    dividendmoney=0
    if (cur_day<=1)
    {
       dividend = 0
    }
        
    else
    {
        for (i in 1:length(cur_stock)){
            close = round(universe[cur_day-1,paste(cur_stock[i],'_Close',sep="")]/universe[cur_day,paste(cur_stock[i],'_Close',sep="")],6)
            adjclose = round(universe[cur_day-1,paste(cur_stock[i],'_Adj.Close',sep="")]/universe[cur_day,paste(cur_stock[i],'_Adj.Close',sep="")],6)
            diff = close-adjclose
            if (diff > 0.01 | diff < -0.01){
                dividend[i] = round(diff * universe[cur_day,paste(cur_stock[i],'_Close',sep="")],2)
            } else {
                dividend[i] = 0
            }   
            dividendmoney[i] = shares[i] *dividend[i]
        }        
    }
    return(sum(dividendmoney))
}

# Test
shares = num_shares(cur_stock,cash_account,universe,cur_day)
get_dividend(universe,cur_stock,27,shares)
```

```r
# while loop to update all value until last day of 2018

while(cur_day < length(universe[,1])){
    for (i in 1:days_interval){
        print(cur_day+i-1)
        MTM_series[cur_day+i-1] = calc_mtm(MTM,shares,universe,cur_day+i-1,cur_stock,cash_balance)+get_dividend(universe,cur_stock,cur_day+i-1,shares)
    }
    cur_day = day_change(days_interval,cur_day)
    print(paste('day: ', cur_day))
    if (cur_day > 250){
        cur_day = 250
        print(paste('day: ',cur_day))
        #print(cur_day)
        cash_balance = sell(cur_stock,cash_balance,cur_day,shares,universe)
        print(paste('cash balance after sell: ',cash_balance))
        break
    }
    print('look below')
    for (i in i:days_interval){
        cash_balance = cash_balance + get_dividend(universe,cur_stock,cur_day+i-1,shares)
        print(cash_balance)
    }
    print('also above')
    cash_balance = sell(cur_stock,cash_balance,cur_day,shares,universe)
    print(paste('cash balance after sell: ',cash_balance))
    cur_stock = adj.close_change(stock,cur_day,days_interval,universe)
    print(paste('price drop most: ',cur_stock))
    shares = num_shares(cur_stock,cash_balance,universe,cur_day)
    print(paste('shares:',shares))
    cash_balance = buy_stock(cur_stock,universe,cur_day,shares,cash_balance)
   print(paste('balance after buy: ',cash_balance)) 

}
```
